```python
# Importar as bibliotecas necessárias para esta etapa
import pandas as pd # Apelido 'pd' é uma convenção da comunidade
import matplotlib.pyplot as plt # Apelido 'plt'
import seaborn as sns # Apelido 'sns'

# Definir o caminho para o arquivo de dados
caminho_dados = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'

# Carregar os dados em um DataFrame do Pandas
df = pd.read_csv(caminho_dados)

# 1. Visualizar as 5 primeiras linhas para ter uma noção da estrutura
print("### AMOSTRA DOS DADOS ###")
print(df.head())

# 2. Obter informações sobre os tipos de dados e valores nulos
print("\n### INFORMAÇÕES GERAIS DO DATA FRAME ###")
df.info()

# 3. Obter estatísticas descritivas para as colunas numéricas
print("\n### ESTATÍSTICAS DESCRITIVAS (COLUNAS NUMÉRICAS) ###")
print(df.describe())
```

```python
# Configurar o estilo dos gráficos do seaborn para um visual mais agradável
sns.set(style="whitegrid")

# Criar a figura e o eixo para o gráfico
plt.figure(figsize=(8, 6))

# Criar o gráfico de contagem para a variável 'Churn'
ax = sns.countplot(x='Churn', data=df)

# Adicionar título
ax.set_title('Distribuição de Churn', fontsize=16)

# Adicionar rótulos de porcentagem nas barras
total = len(df['Churn'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_height() + 20
    ax.annotate(percentage, (x, y))

# Exibir o gráfico
plt.show()
```

```python
# Criar um gráfico de contagem para a variável 'Contract', separando por 'Churn'
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Contract', hue='Churn', data=df)

# Adicionar título
ax.set_title('Churn por Tipo de Contrato', fontsize=16)

# Exibir o gráfico
plt.show()
```
